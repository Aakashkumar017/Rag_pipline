```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter=RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunk=splitter.split_documents(doc)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
try:
    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )
    print("HuggingFaceEmbeddings model is scesfully load for embedding ")
except Exception :
    print("Model loading is fail")
```

```python
vector_store=FAISS.from_documents(chunk,embeddings)
```

```python
#Here i am loading my grok model 
from dotenv import load_dotenv
import os 
load_dotenv()
try:
    model=ChatOpenAI(
    model="llama-3.1-8b-instant",
    base_url="https://api.groq.com/openai/v1",
    api_key=os.getenv("GROQ_API_KEY"))

    print("Grok model is sucessfully load")

except Exception as e:
    print (f"Error : {e}")
```

```python
retriever = vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 3,"fetch_k":10,"lambda_mult": 0.7}
)

def retrieve(query):
    return retriever.invoke(query)
```

```python
score = vector_store.similarity_search_with_score("machine learning", k=3)
score
```

```python
for document , score in score:
    print("Score:", float(score))
    print("Content:", document.page_content[:100])
    print("------")
```

```python
def rag_pipeline(query):
    try:
        doc_retriver= retrieve(query)
        if not doc_retriver:
            return "No relevant information found in document."
        context = "\n\n".join([
            f"[Source: Page {doc.metadata.get('page')} | File: {doc.metadata.get('source')}] \n{doc.page_content}"
            for doc in doc_retriver
        ])
        prompt = f"""
            You are a helpful chatbot. Answer only from the context below.
            If the answer is not in context, say "I don't know".
            RULES:
                - Answer ONLY from the provided context
                - If answer is not explicitly present → say "I don't know"

            Context:
            {context}

            Question:
            {query}

            Give a clear, structured answer.
        """
        result= model.invoke(prompt)
        return result.content  
    except Exception as e:
        print(f"error: {e}")
```

```python

# print(rag_pipeline("pm of india"))
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

chat_history = [
    SystemMessage(content="You are an AI that answers questions from PDF.")
]

while True:
    user_input = input("You: ")

    if not user_input.strip():
        print("Waiting for input...")
        continue

    chat_history.append(HumanMessage(content=user_input))

    # 🔥 Only pass latest query to RAG
    result = rag_pipeline(user_input)

    chat_history.append(AIMessage(content=result))

    print("AI:", result)
```
